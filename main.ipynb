```python
import pandas as pd
import os
```

# 1. Limpeza e formatação dos arquivos

```python
def clean_file(filename: str) -> None:
    data = pd.read_csv(
        f'./files/raw/{filename}',
        # engine='pyarrow',
        dtype=str
    )
    
    # remove espaços em branco
    # remove colunas vazias
    # remove linhas vazias
    data = (
        data
        .apply(lambda v: v.str.replace(r"\s+", " ", regex=True).str.strip())
        .dropna(axis=1, how='all')
        .dropna(axis=0, how='all')
    )
    
    while '' in data.columns:
        data = data.drop(columns='')
    
    for col in data.columns:
        if 'Unnamed' in col:
            data = data.drop(columns=col)
    
    # salva como um novo CSV
    data.to_csv(f'./files/cleaned/{filename}', index=False)
    
    print(f'Arquivo {filename} limpo! (linhas x colunas: {data.shape})')
    if data.shape[1] > 30:
        print('Colunas:', data.columns)
```

```python
files = sorted(os.listdir('./files/raw'))[7:]

for filename in files:
    clean_file(filename)
```
